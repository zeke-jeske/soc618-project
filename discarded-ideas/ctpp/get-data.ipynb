```python
import pandas as pd
import requests
from sympy import resultant

api_key = 'c7y8f1seygk9atf4r1dan2bz'
api_base_url = 'https://ctppdata.transportation.org/api/'
```

```python
default_headers = {
    'accept': 'application/json',
    'x-api-key': api_key
}
default_params = {}
def fetch_api(endpoint, headers = default_headers, params = default_params):
    url = api_base_url + endpoint
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Request failed with status code {response.status_code}")
        raise Exception("API request failed")
```
